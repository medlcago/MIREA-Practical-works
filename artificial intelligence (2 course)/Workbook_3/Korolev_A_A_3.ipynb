**1.2.1. –ü—Ä–∏–º–µ—Ä**

–ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é sum_range(start, end), –∫–æ—Ç–æ—Ä–∞—è —Å—É–º–º–∏—Ä—É–µ—Ç –≤—Å–µ —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞ –æ—Ç –∑–Ω–∞—á–µ–Ω–∏—è ¬´start¬ª –¥–æ –≤–µ–ª–∏—á–∏–Ω—ã ¬´end¬ª –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ.
–ï—Å–ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –∑–∞–¥–∞—Å—Ç –ø–µ—Ä–≤–æ–µ —á–∏—Å–ª–æ –±–æ–ª—å—à–µ–µ —á–µ–º –≤—Ç–æ—Ä–æ–µ, –ø—Ä–æ—Å—Ç–æ –ø–æ–º–µ–Ω—è–π—Ç–µ –∏—Ö –º–µ—Å—Ç–∞–º–∏.

```python
def sum_range(start, end):
    if start > end:
        end, start = start, end
    return sum(range(start, end + 1))


print(sum_range(2, 12))
print(sum_range(-4, 4))
print(sum_range(3, 2))
```

**1.2.2. –ü—Ä–∏–º–µ—Ä**

–ù–∞–ø–∏—à–∏—Ç–µ —Ä–µ–∫—É—Ä—Å–∏–≤–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é –≤—ã—á–∏—Å–ª–µ–Ω–∏—è —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª–∞ –Ω–∞ —è–∑—ã–∫–µ Python.

```python
def fact(num):
    if num == 0:
        return 1
    else:
        return num * fact(num - 1)


print(fact(5))
```

–°—Ö–æ–¥—Å—Ç–≤–æ –∏–ª–∏ —Ä–∞–∑–ª–∏—á–∏–µ –º–µ–∂–¥—É –æ–±—ä–µ–∫—Ç–∞–º–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏

—Ç–æ –æ–Ω –º–æ–∂–µ—Ç –±—ã—Ç—å –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω –∫–∞–∫ —Ç–æ—á–∫–∞ –≤ n-–º–µ—Ä–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ, –∏
—Å—Ö–æ–¥—Å—Ç–≤–æ —Å –¥—Ä—É–≥–∏–º–∏ –æ–±—ä–µ–∫—Ç–∞–º–∏ –±—É–¥–µ—Ç –æ–ø—Ä–µ–¥–µ–ª—è—Ç—å—Å—è –∫–∞–∫ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–µ
—Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ. –ü—Ä–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –∏—Å–ø–æ–ª—å–∑—É—é—Ç—Å—è —Ä–∞–∑–ª–∏—á–Ω—ã–µ –º–µ—Ä—ã —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è
–º–µ–∂–¥—É –æ–±—ä–µ–∫—Ç–∞–º–∏.

1.	*–ï–≤–∫–ª–∏–¥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ*

–≠—Ç–æ, –ø–æ–∂–∞–ª—É–π, –Ω–∞–∏–±–æ–ª–µ–µ —á–∞—Å—Ç–æ –∏—Å–ø–æ–ª—å–∑—É–µ–º–∞—è –º–µ—Ä–∞ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è. –û–Ω–∞ —è–≤–ª—è–µ—Ç—Å—è –≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–º —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ–º –≤ –º–Ω–æ–≥–æ–º–µ—Ä–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ –∏ –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º:

$P = \sqrt{\displaystyle\sum_{i=1}^{N} (A_i - B_i)^2}$

–≥–¥–µ

    ùëÉ ‚Äì —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –æ–±—ä–µ–∫—Ç–∞–º–∏ A –∏ B;
    ùê¥ùëñ ‚Äì –∑–Ω–∞—á–µ–Ω–∏–µ i-—Å–≤–æ–π—Å—Ç–≤–∞ –æ–±—ä–µ–∫—Ç–∞ A;
    ùêµùëñ ‚Äì –∑–Ω–∞—á–µ–Ω–∏–µ i-—Å–≤–æ–π—Å—Ç–≤–∞ –æ–±—ä–µ–∫—Ç–∞ B.

2. *–ö–≤–∞–¥—Ä–∞—Ç –µ–≤–∫–ª–∏–¥–æ–≤–∞ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è*

–î–∞–Ω–Ω–∞—è –º–µ—Ä–∞ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –≤ —Ç–µ—Ö —Å–ª—É—á–∞—è—Ö, –∫–æ–≥–¥–∞ —Ç—Ä–µ–±—É–µ—Ç—Å—è
–ø—Ä–∏–¥–∞—Ç—å –±–æ–ª—å—à–µ –∑–Ω–∞—á–µ–Ω–∏–µ –±–æ–ª–µ–µ –æ—Ç–¥–∞–ª–µ–Ω–Ω—ã–º –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞ –æ–±—ä–µ–∫—Ç–∞–º. –≠—Ç–æ
—Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º:

$P = \displaystyle\sum_{i=1}^{N} (A_i - B_i)^2$

3. *–í–∑–≤–µ—à–µ–Ω–Ω–æ–µ –µ–≤–∫–ª–∏–¥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ*

–ü—Ä–∏–º–µ–Ω—è–µ—Ç—Å—è –≤ —Ç–µ—Ö —Å–ª—É—á–∞—è—Ö, –∫–æ–≥–¥–∞ –∫–∞–∂–¥–æ–º—É i-—Å–≤–æ–π—Å—Ç–≤—É —É–¥–∞–µ—Ç—Å—è –ø—Ä–∏–ø–∏—Å–∞—Ç—å
–Ω–µ–∫–æ—Ç–æ—Ä—ã–π ¬´–≤–µ—Å¬ª wi, –ø—Ä–æ–ø–æ—Ä—Ü–∏–æ–Ω–∞–ª—å–Ω–æ —Å—Ç–µ–ø–µ–Ω–∏ –≤–∞–∂–Ω–æ—Å—Ç–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞ –≤ –∑–∞–¥–∞—á–µ
–∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏:

$P = \sqrt{\displaystyle\sum_{i=1}^{N} w_i(A_i - B_i)^2}$

4. *–•–µ–º–º–∏–Ω–≥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ*

–¢–∞–∫–∂–µ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –º–∞–Ω—Ö—ç—Ç—Ç–µ–Ω—Å–∫–∏–º, —Å–∏—Ç–∏-–±–ª–æ–∫ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ–º –∏–ª–∏ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ–º
–≥–æ—Ä–æ–¥—Å–∫–∏—Ö –∫–≤–∞—Ä—Ç–∞–ª–æ–≤. –≠—Ç–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ —è–≤–ª—è–µ—Ç—Å—è —Ä–∞–∑–Ω–æ—Å—Ç—å—é –ø–æ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º.
–•–µ–º–º–∏–Ω–≥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è –ø–æ —Ñ–æ—Ä–º—É–ª–µ:

$P = \displaystyle\sum_{i=1}^{N} (|A_i| - |B_i|)$

5. *–†–∞—Å—Å—Ç–æ—è–Ω–∏–µ –ß–µ–±—ã—à–µ–≤–∞*

–ü—Ä–∏–Ω–∏–º–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ –Ω–∞–∏–±–æ–ª—å—à–µ–≥–æ –º–æ–¥—É–ª—è —Ä–∞–∑–Ω–æ—Å—Ç–∏ –º–µ–∂–¥—É –∑–Ω–∞—á–µ–Ω–∏—è–º–∏
—Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö —Å–≤–æ–π—Å—Ç–≤ (–ø—Ä–∏–∑–Ω–∞–∫–æ–≤) –æ–±—ä–µ–∫—Ç–æ–≤:

$P = \max(|A_i - B_i|)$

–í—ã–±–æ—Ä –º–µ—Ä—ã —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –∏ –≤–µ—Å–æ–≤ –¥–ª—è –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä—É—é—â–∏—Ö —Å–≤–æ–π—Å—Ç–≤ ‚Äì
–æ—á–µ–Ω—å –≤–∞–∂–Ω—ã–π —ç—Ç–∞–ø, —Ç–∞–∫ –∫–∞–∫ –æ—Ç —ç—Ç–∏—Ö –ø—Ä–æ—Ü–µ–¥—É—Ä –∑–∞–≤–∏—Å—è—Ç —Å–æ—Å—Ç–∞–≤ –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ
—Ñ–æ—Ä–º–∏—Ä—É–µ–º—ã—Ö –∫–ª–∞—Å—Å–æ–≤, –∞ —Ç–∞–∫–∂–µ —Å—Ç–µ–ø–µ–Ω—å —Å—Ö–æ–¥—Å—Ç–≤–∞ –æ–±—ä–µ–∫—Ç–æ–≤ –≤–Ω—É—Ç—Ä–∏ –∫–ª–∞—Å—Å–æ–≤.

**1.2.3. –ü—Ä–∏–º–µ—Ä**

–ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏–∏ –≤ Python, –∫–æ—Ç–æ—Ä–∞—è –≤—ã—á–∏—Å–ª—è–µ—Ç –ï–≤–∫–ª–∏–¥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –¥–≤—É–º—è –º–∞—Å—Å–∏–≤–∞–º–∏ NumPy.

```python
import numpy as np


def euclidean_distance(v1, v2):
    return sum((x - y) ** 2 for x, y in zip(v1, v2)) ** 0.5


x = np.array([0, 0, 0])
y = np.array([3, 3, 3])
print(euclidean_distance(x, y))
```

**1.2.4. –ü—Ä–∏–º–µ—Ä**

–ù–∞–ø–∏—à–∏—Ç–µ 4 —Ñ—É–Ω–∫—Ü–∏–∏ –≤ Python, –∫–æ—Ç–æ—Ä—ã–µ —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞—é—Ç –∫–≤–∞–¥—Ä–∞—Ç –ï–≤–∫–ª–∏–¥–æ–≤–∞ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è, –≤–∑–≤–µ—à–µ–Ω–Ω–æ–µ –µ–≤–∫–ª–∏–¥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ, –•–µ–º–º–∏–Ω–≥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –∏ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –ß–µ–±—ã—à–µ–≤–∞ –º–µ–∂–¥—É –¥–≤—É–º—è –º–∞—Å—Å–∏–≤–∞–º–∏ NumPy.

```python
def sqr_euclidean_distance(v1, v2):
    return sum((x - y) ** 2 for x, y in zip(v1, v2))


def weighted_euclidean_distance(v1, v2, w):
    return sum((x - y) ** 2 * s for x, y, s in zip(v1, v2, w)) ** 0.5


def manhattan_distance(v1, v2):
    return sum(abs(x - y) for (x, y) in zip(v1, v2))


def chebyshev_distance(v1, v2):
    return max(abs(x - y) for (x, y) in zip(v1, v2))


x = np.array([0, 0, 0])
y = np.array([3, 3, 3])
w = np.array([0, 0, 1])
print(sqr_euclidean_distance(x, y))
print(weighted_euclidean_distance(x, y, w))
print(manhattan_distance(x, y))
print(chebyshev_distance(x, y))
```

**1.3.1. –ó–∞–¥–∞–Ω–∏–µ**

–ó–∞–¥–∞–π—Ç–µ 4 —Ç–æ—á–∫–∏ –≤ —Ç—Ä–µ—Ö–º–µ—Ä–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ, —Ä–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ –º–µ–∂–¥—É –Ω–∏–º–∏ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –ø–æ –æ–ø–∏—Å–∞–Ω–Ω—ã–º –≤ –ø—Ä–∏–º–µ—Ä–µ –≤—ã—à–µ –º–µ—Ç—Ä–∏–∫–∞–º. –û—Ç–æ–±—Ä–∞–∑–∏—Ç–µ —Ç–æ—á–∫–∏ –≤ —Ç—Ä–µ—Ö–º–µ—Ä–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ.

```python
import itertools

P1 = np.array([1, 2, 3])
P2 = np.array([4, 5, 6])
P3 = np.array([-1, -2, -3])
P4 = np.array([-4, -5, -6])

points = [P1, P2, P3, P4]
point_labels = ["P", "P2", "P3", "P4"]

weights = np.array([0.5, 1, 1.5])

for pair in itertools.combinations(range(len(points)), 2):
    A, B = points[pair[0]], points[pair[1]]
    sq_euclidean_dist = np.linalg.norm(A - B) ** 2
    weighted_euclidean_dist = np.linalg.norm(weights * (A - B)) ** 2
    hamming_dist = np.linalg.norm(A != B, ord=0)
    chebyshev_dist = np.linalg.norm(A - B, ord=np.inf)

    print(f"Distances between {point_labels[pair[0]]} and {point_labels[pair[1]]}:")
    print(f"\tSquared Euclidean distance: {sq_euclidean_dist:.2f}")
    print(f"\tWeighted Euclidean distance: {weighted_euclidean_dist:.2f}")
    print(f"\tHamming distance: {hamming_dist:.2f}")
    print(f"\tChebyshev distance: {chebyshev_dist:.2f}\n")

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

for point, label in zip(points, point_labels):
    ax.scatter(point[0], point[1], point[2], label=label)
    ax.text(point[0], point[1], point[2], label)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.legend()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

# –¥–∞–Ω–Ω—ã–µ
X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
target = [0, 0, 0, 1, 1, 1]

# –æ–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å k-–±–ª–∏–∂–∞–π—à–∏—Ö —Å–æ—Å–µ–¥–µ–π –∫ –¥–∞–Ω–Ω—ã–º
K = 3
model = KNeighborsClassifier(n_neighbors=K)
model.fit(X, target)
print(model)

# –¥–µ–ª–∞–µ–º –ø—Ä–æ–≥–Ω–æ–∑
print('(-2,-2) is class'),
print(model.predict([[-2, -2]]))

print('(1,3) is class'),
print(model.predict([[1, 3]]))
```

**2.3.1. –ó–∞–¥–∞–Ω–∏–µ**

–î–ª—è –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ –ø—Ä–∏–º–µ—Ä–∞ –ø–æ—ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∏—Ä—É–π—Ç–µ —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞:
1.	–£—Å—Ç–∞–Ω–æ–≤–∏—Ç–µ –¥—Ä—É–≥–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–ª–∏–∂–∞–π—à–∏—Ö —Å–æ—Å–µ–¥–µ–π (k = 1, 5, 10).
2.	–£—Å—Ç–∞–Ω–æ–≤–∏—Ç–µ —Ä–∞–∑–º–µ—Ä —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏ 15% –æ—Ç –≤—Å–µ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞.
3.	–ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫–∏ –∏ –æ—Ü–µ–Ω–∏—Ç–µ –∫–∞—á–µ—Å—Ç–≤–æ –º–æ–¥–µ–ª–µ–π, –ø—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä—É–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã.


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target
iris_df['species'] = iris_df['species'].map({0: iris.target_names[0], 1: iris.target_names[1], 2: iris.target_names[2]})


def exp3(k_neighbors=3, test_size_ratio=0.2):
    X_train, X_test, y_train, y_test = train_test_split(
        iris_df.iloc[:, :-1],
        iris_df.iloc[:, -1],
        test_size=test_size_ratio
    )

    model = KNeighborsClassifier(n_neighbors=k_neighbors)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    plt.figure(figsize=(10, 7))
    sns.scatterplot(x='petal width (cm)', y='petal length (cm)', data=iris_df, hue='species', s=70)
    plt.xlabel("–î–ª–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–∞, —Å–º")
    plt.ylabel("–®–∏—Ä–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–∞, —Å–º")
    plt.legend(loc=2)
    plt.grid()

    for i in range(len(y_test)):
        if np.array(y_test)[i] != y_pred[i]:
            plt.scatter(X_test.iloc[i, 3], X_test.iloc[i, 2], color='red', s=150)

    print(f'k_neighbors: {k_neighbors}, test_size_ratio: {test_size_ratio}')
    print(f'accuracy: {accuracy_score(y_test, y_pred):.3}')


# –≠–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç—ã —Å —Ä–∞–∑–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ k_neighbors –∏ test_size_ratio
exp3(k_neighbors=1, test_size_ratio=0.15)
plt.show()

exp3(k_neighbors=5, test_size_ratio=0.15)
plt.show()

exp3(k_neighbors=10, test_size_ratio=0.15)
plt.show()
```

1. –ü—Ä–∏ k_neighbors=1, –º–æ–¥–µ–ª—å –∏–º–µ–µ—Ç –≤—ã—Å–æ–∫—É—é —Ç–æ—á–Ω–æ—Å—Ç—å (–≤ –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ 1.0), –Ω–æ —Å—Ç–æ–∏—Ç –æ—Ç–º–µ—Ç–∏—Ç—å, —á—Ç–æ –º–æ–¥–µ–ª—å —Å –æ–¥–Ω–∏–º –±–ª–∏–∂–∞–π—à–∏–º —Å–æ—Å–µ–¥–æ–º –º–æ–∂–µ—Ç –±—ã—Ç—å —á—Ä–µ–∑–º–µ—Ä–Ω–æ —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–∞ –∫ –≤—ã–±—Ä–æ—Å–∞–º –∏ —à—É–º—É, —á—Ç–æ –º–æ–∂–µ—Ç –ø—Ä–∏–≤–µ—Å—Ç–∏ –∫ –ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∏—é –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ.
2. –ü—Ä–∏ k_neighbors=5, –º–æ–¥–µ–ª—å —Ç–∞–∫–∂–µ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –æ—Ç–ª–∏—á–Ω—É—é —Ç–æ—á–Ω–æ—Å—Ç—å (1.0).
3. –ü—Ä–∏ k_neighbors=10, –º–æ–¥–µ–ª—å –≤—Å–µ –µ—â–µ –¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É–µ—Ç –ø—Ä–∏–µ–º–ª–µ–º—É—é —Ç–æ—á–Ω–æ—Å—Ç—å (0.957), –Ω–æ –Ω–∞–±–ª—é–¥–∞–µ—Ç—Å—è –Ω–µ–∫–æ—Ç–æ—Ä–æ–µ —Å–Ω–∏–∂–µ–Ω–∏–µ –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –ø–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏—é —Å k_neighbors=5. –≠—Ç–æ –æ–±—ä—è—Å–Ω—è–µ—Ç—Å—è —Ç–µ–º, —á—Ç–æ —Å —É–≤–µ–ª–∏—á–µ–Ω–∏–µ–º —á–∏—Å–ª–∞ —Å–æ—Å–µ–¥–µ–π –º–æ–¥–µ–ª—å —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –º–µ–Ω–µ–µ —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ–π –∫ –ª–æ–∫–∞–ª—å–Ω—ã–º –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç—è–º –¥–∞–Ω–Ω—ã—Ö –∏ –º–æ–∂–µ—Ç –ø—Ä–æ–ø—É—Å—Ç–∏—Ç—å –Ω–µ–∫–æ—Ç–æ—Ä—ã–µ –Ω—é–∞–Ω—Å—ã.

# **–†–∞–±–æ—Ç–∞ —Å –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏**
