# **Типы данных**

        списки (list),

# **If - Условный оператор**

В коде часто приходится проверять выполнимость или невыполнимость каких-то условий.

Синтаксис:

    if<условие1 (булевское выражение)> :
        <код, который выполнится, если условие верно>
    elif<условие2 (булевское выражение)>:
        <код, который выполнится, если условие1 было неверно, а условие2верно>
    else:
        <код, который выполнится, если условие1 и условие2 были неверны>
Обратите внимание, что код, который должен выполняться внутри каждого условия, записывается с отступом в 4 пробела от уровня if, elif и else: в питоне области видимости переменных обозначаются отступами.

То есть, отступы позволяют понять, где начинается код, который должен выполняться при выполнении условия в if, и где заканчивается.

**2.2. Пример**

Вывести на экран является ли переменная х положительной, отрицательной или равна нулю.

```python
x = 125
if x < 0:
    print("x отрицательный")
elif x == 0:
    print("x равен нулю")
else:
    print("x положительный")
```

# **Циклы - for и while**

Цикл while проверяет истинность условия, и если условие истинно, то выполняет инструкции цикла.

Синтаксис:

    while условие:
      инструкции
Инструкции, которые относятся к циклу while, располагаются на последующих строках и должны иметь отступ от начала ключевого слова while.
    
Цикл for проходит по набору значений, помещает каждое значение в переменную с которой в цикле можно производить различные действия.

Синтаксис:
    
    for переменная in набор_значений:
        инструкции
После ключевого слова for идет название переменной, в которую будут помещаться значения. После оператора in указывается набор значений и двоеточие. Инструкции, которые относятся к циклу, располагаются на последующих строках и должны иметь отступ.
При создании цикла удобно пользоваться функцией range(a, b), которая создает последовательность чисел от a до b-1

**3.2.1. Пример**

**4.2.2. Пример**

Для функции $$\sin(x \cdot e^{\cos(x)})$$
 на интервале [0;10] с шагом 0,05 построить график функции, закрасить площадь под ним и вычислить значение этой площади. Для вычисления площади воспользуйте функции trapz(), вычисляющей интеграл по правилу трапеции. Для ее корректной работы необходимо подключить следующие библиотеки: from scipy.integrate import simps from numpy import trapz

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import trapz

x = np.arange(0.0, 10, 0.1)
y = np.abs(np.sin(x * np.exp(np.cos(x))))
plt.grid()
plt.plot(x, y, c="r")
plt.fill_between(x, y)
plt.show()

area = trapz(y)
print(area)
```

**4.3.2. Задание**

Дана функция $$\frac{\sqrt{1 + e^{\sqrt{x}} + \cos(x^2)}}{\left|1 - \sin^3(x)\right|} + \ln\left(\left|2x\right|\right)$$
Создать массив из 10 значений функции (x, например, изменяется от 1 до 10). Выделить срез первой половины массива и построить графики для основного массива, используя plt.plot(), для среза – точечный (plt.scatter()).

```python
import numpy as np
from matplotlib import pyplot as plt


def func(x):
    return (np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(x ** 2)) / np.abs(1 - np.sin(x) ** 3)) + np.log(np.abs(2 * x))


x_values = np.linspace(1, 10, 10)
y_values = func(x_values)

x_slice = x_values[:5]
y_slice = y_values[:5]

plt.figure(figsize=(10, 5))
plt.plot(x_values, y_values, label="Основной массив")
plt.scatter(x_slice, y_slice, color="red", label="Срез первой половины")
plt.xlabel("x")
plt.ylabel("y")
plt.title("График функции и срез первой половины")
plt.legend()
plt.grid()
plt.show()
```

**4.3.3. Задание**

Дана функция $$\left|\cos\left(x \cdot e^{\left(\cos(x) + \ln(x + 1)\right)}\right)\right|$$
Построить график на интервале (0, 10) с шагом 1 с заливкой площади и найти эту площадь под ним. Для вычисления площади воспользуйтесь функцией trapz(), вычисляющей интеграл по правилу трапеции.
Для ее корректной работы необходимо подключить следующие библиотеки:
    
    from scipy.integrate import simps
    from numpy import trapz

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import simps


def func(x):
    return np.abs(np.cos(x * np.exp(np.cos(x) + np.log(x + 1))))


x_values = np.linspace(1, 10, 10)
y_values = func(x_values)

plt.figure(figsize=(10, 5))
plt.plot(x_values, y_values, label="Функция", marker='o')
plt.fill_between(x_values, y_values, alpha=0.2)
plt.xlabel("x")
plt.ylabel("y")
plt.title("График функции и площадь под ним")
plt.legend()
plt.grid()
plt.show()

area = simps(y_values, x_values)
print(f"Площадь под графиком методом Симпсона: {area:.3f}")
area = np.trapz(y_values, x_values)
print(f"Площадь под графиком методом трапеции: {area:.3f}")
```

**4.3.4. Задание**

Откройте статистические данные по стоимости акций Apple, Microsoft, Google за 12 месяцев 2021 г. Постройте 3 графика на плоскости и оцените их динамику. Подумайте, как можно улучшить читаемость результатов.

```python
import yfinance as yf
import datetime
import matplotlib.pyplot as plt


def get_stock_data(ticker, start, end):
    return yf.download(ticker, start=start, end=end)


def plot_stock_prices_overlay(stock_data, title):
    plt.figure(figsize=(20, 10))
    for ticker, data, color in stock_data:
        data["Close"].plot(label=ticker, color=color, grid=True)
    plt.xlabel("Дата")
    plt.ylabel("Цена закрытия")
    plt.title(title)
    plt.legend()
    plt.show()


start_date = datetime.datetime(2021, 1, 1)
end_date = datetime.datetime(2022, 1, 1)

apple_data = get_stock_data("AAPL", start_date, end_date)
microsoft_data = get_stock_data("MSFT", start_date, end_date)
google_data = get_stock_data("GOOGL", start_date, end_date)

stock_data = [
    ("Apple", apple_data, "tab:blue"),
    ("Microsoft", microsoft_data, "tab:green"),
    ("Google", google_data, "tab:red")
]

plot_stock_prices_overlay(stock_data, "Цены на акции Apple, Microsoft и Google (2021 г.)")
```

**4.3.5. Задание**

Создайте простейший калькулятор, включающий основные действия для двух переменных '+', '-', '∙', '÷', а также вычисление следующих функций:

$e^{x+y}$

$\sin(x + y)$

$\cos(x + y)$

$x^y$

```python
import math


def calculator(operation, x, y):
    if operation == "exp":
        return math.exp(x + y)
    elif operation == "pow":
        return x ** y
    elif operation == "sin":
        return math.sin(x + y)
    elif operation == "cos":
        return math.cos(x + y)
    else:
        raise ValueError("Неизвестная операция")


x = float(input("Введите x: "))
y = float(input("Введите y: "))

print("Выберите операцию:")
print("1. e^(x+y)")
print("2. x^y")
print("3. sin(x + y)")
print("4. cos(x + y)")

operation = input("Введите номер операции (1-4): ")

if operation == "1":
    result = calculator("exp", x, y)
elif operation == "2":
    result = calculator("pow", x, y)
elif operation == "3":
    result = calculator("sin", x, y)
elif operation == "4":
    result = calculator("cos", x, y)
else:
    raise ValueError("Неверный номер операции")

print(f"Ответ: {result}")
```
